# *Load Dataset*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Define column names
column_names = [
    'ID', 'Age', 'Gender', 'Education', 'Country', 'Ethnicity', 'Nscore', 'Escore', 
    'Oscore', 'Ascore', 'Cscore', 'Impulsive', 'SS', 'Alcohol', 'Amphet', 'Amyl', 
    'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy', 'Heroin', 
    'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms', 'Nicotine', 'Semer', 'VSA'
]

# Load the dataset
df = pd.read_csv('drug_consumption.data', header=None, names=column_names)

# Check the first few rows
print(df.head())
```

# *Basic Explorations*

```python
print(df.shape)  # check dataset shape
```

```python
print(df.info()) #data types
```

```python
print(df.describe()) #summary statisitcs
```

# *Explore Catergorical Variables*

```python
print(df['Gender'].unique())  # Example for gender
```

```python

# Example: Gender distribution
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('gender_distribution.png')
plt.show()
```

# *Explore Numerical Variables*

```python
# Example: Neuroticism score distribution (Histogram)
df['Nscore'].hist(bins=20)
plt.title('Distribution of Neuroticism Score')
plt.xlabel('Neuroticism Score')
plt.ylabel('Frequency')
plt.savefig('neuroticism_distribution.png')
plt.show()
```

```python
# Plot neuroticism score by gender  (Box Plots)
df.boxplot(column='Nscore', by='Gender')
plt.title('Neuroticism Score by Gender')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Gender')
plt.ylabel('Neuroticism Score')

# Save the plot
plt.savefig('neuroticism_by_gender.png')  # Saves as PNG file
plt.show()  # Optionally display the plot
```

# *Explore Target Variables*

```python
print(df['Cannabis'].value_counts())
```

```python
df['Cannabis'].value_counts().plot(kind='bar')
plt.title('Cannabis Use')
plt.xlabel('Cannabis Use')
plt.ylabel('Count')
plt.savefig('Cannabis use.png')
plt.show()
```

# *Explore relationship between variables*

```python

# Calculate correlation matrix
corr = df[['Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore']].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('Correlation matrix.png')
plt.show()
```

# **Grouped analysis**

```python
# Example: Cannabis use by gender
print(df.groupby('Gender')['Cannabis'].value_counts(normalize=True))
```

# *Check for outliers*

```python
# Box plot for all numerical columns
numerical_columns = ['Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore', 'Impulsive', 'SS']
df[numerical_columns].plot(kind='box', figsize=(12, 6), title='Box Plot of Numerical Columns')
plt.savefig('numerical_boxplots.png')
plt.show()
```
